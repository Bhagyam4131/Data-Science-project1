```python
import matplotlib.pyplot as plt
categories =["transport", "income tax","housing"]
values1 =[15,24,30]
values2 = [20,18,25]
plt.bar(categories,values1,label="first half",color="skyblue")
plt.bar(categories,values2, bottom=values1, label="second half",color="pink")
plt.xlabel("categories")
plt.ylabel("values")
plt.title("stacked bar graph")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
mark1 = np.array([2,4,5,7,9,10,11,13,15,20])
mark2 = np.array([6,7,8,10,11,12,13,14,15,16])
plt.hist([mark1,mark2],bins=10,stacked=True,color=["skyblue","salmon"],edgecolor="black")
plt.xlabel("marks")
plt.ylabel("count")
plt.title("stacked histogram with multiple datasets")
plt.legend(["mark dataset1", "mark dataset2"])
plt.show()
```

```python
#pandas library
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a)
print(a)
print(myvar)
print('..................')
#labels
print(myvar[0])
print('........................')
myvar2=pd.Series(a, index =["x","y","z"])
print(myvar2)
print(myvar2["y"])
print('......................')
calories={"day1":420, "day2":380, "day3":390}
myvar=pd.Series(calories)
print(myvar)
print('........................')
myvar = pd.Series(calories,index =["day1","day2"])
print(myvar)
print('.......................')
series = pd.Series(5, index=[0,1,2,3])
print(series)
print('..........................')
```

```python
#integer Series
int_series = pd.Series([1,2,3,4])
print(int_series)
print('....................')
#float Series
float_series =pd.Series([1.1,2.2,3.3,4.4])
print(float_series)
print('...................')
#string Series
string_series = pd.Series(['a','b','c','d'])
print(string_series)
print('.......................')
#label-based indexing
data ={'a':10,'b':20,'c':30}
series=pd.Series(data)
print(series['a'])
print(series[['a','c']])
print('.......................')
#boolean inexing
data = [10,20,30,40,50]
series = pd.Series(data)
#access element greater than 25
print(series[series > 25])
print('...............................')
```

```python
#Data frame
data ={
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,30,35],
    'City':['New York','Los Angeles','Chicago']
}
df = pd.DataFrame(data)
print(df)
print('...........................')
#selecting columns
print(df.Name)
print(df['Name'])
print(df[['Name', 'City']])
print('....................')
print(df.loc[0])
print(df.iloc[0])
#select multiple rows
print(df.iloc[0:2])
print('................')

```

```python
#conditional selection
data ={
    'name':['alice','bob','charlie'],
    'age':[25,30,35],
    'city':['new york','los angeles','chicago']
}
df=pd.DataFrame(data)
print(df[df['age'] > 25])
print('..............')
print(df.loc[(df.name=="bob") &(df.city=="los angeles")])
```

```python
#Grouping by a single column
data ={
    'name':['alice','bob','charlie','alice'],
    'age':[25,30,35,25],
    'city':['new york','san francisco','new york','san francisco'],
    'salary':[70000,80000,120000,70000]
}
df=pd.DataFrame(data)
grouped = df.groupby('name')
print(grouped.groups)
#calculate the sum of salaries for each group
salary_sum=grouped['salary'].sum()
print(salary_sum)
age_mean=grouped['age'].mean()
print(age_mean)
#grouping by multiple columns
#group by 'name' and 'city'
grouped_multiple=df.groupby(['name','city'])
#calculatemthe sum of salaries for each group
salary_sum_multiple=grouped_multiple['salary'].sum()
print(salary_sum_multiple)
```

```python
import pandas as pd
#define a dictionary containing employee data
data1={'name':['jai','prinaci','gaurav','anuj'],
       'age':['27','30','42','42'],
        'address':['nagpur','kanpur','hyd','guntur'],
        'Qualification': ['msc','ma','mca','mba']}
data2={'name':['bhagya','prameela','sam','ramya'],
       'age':['19','18','20','21'],
        'address':['mch','mcl','ongole','nrt'],
        'Qualification':['Btech','MS','Bcom','Bsc']}
#convert  the dictionary into dataframe
df1=pd.DataFrame(data1,index=[0,1,2,3])
df2=pd.DataFrame(data2,index=[4,5,6,7])
print(df1)
print(df2)
print('.......................')

#applying concat with axes
#join = 'inner' we get the intersection of dataframe
res2= pd.concat([df1,df2], axis=1, join='inner')
print(res2)
print('.............................')

#covert the dictionary into DataFrame
df1=pd.DataFrame(data1,index=[0,1,2,3])
df2=pd.DataFrame(data2,index=[2,3,6,7])
print(df1, "\n\n",df2)
print('...............')
#using ignore_index
df1=pd.DataFrame(data1,index=[0,1,2,3])
df2=pd.DataFrame(data2,index=[4,5,6,7])
res = pd.concat([df1,df2], ignore_index=True)
print(res)
print('....................')
#convert the dictionary into DataFrame
df1=pd.DataFrame(data1,index=[0,1,2,3])
#creating a series
s1=pd.Series([1000,2000,3000,4000],name='Salary')
print(df1,"\n\n",s1)
res = pd.concat([df1,s1], axis=1)
print(res)
print('..........................')
#merging dataframes
#define a dictionary containing employee data
data1 ={'key':['k0','k1','k2','k3'],
        'name':['jai','princi','kavya','anuj'],
        'age':[27,28,22,24],}
data2 ={'key':['k0','k1','k2','k3'],
        'name':['bhagya','prameela','sam','ramya'],
        'qualification':['btech','ms','bcom','bsa'],}
df1=pd.DataFrame(data1,index=[0,1,2,3])
df2=pd.DataFrame(data2,index=[4,5,6,7])
res = pd.merge(df1,df2, on='key')
print(res)
print('..................')

data1 ={'key':['k0','k1','k2','k3'],
        'key1':['k0','k1','k0','k1'],
        'name':['jai','princi','kavya','anuj'],
        'age':[27,28,22,24],}
data2 ={'key':['k0','k1','k2','k3'],
        'key1':['k0','k0','k0','k0'],
        'name':['bhagya','prameela','sam','ramya'],
        'qualification':['btech','ms','bcom','bsa'],}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
print(df1,"\n\n",df2)
res1=pd.merge(df1,df2, on=['key', 'key1'])
print(res1)
```
